## Poiseuille's Law

Poiseuille's Law states that the flow rate Q through a vessel is proportional to the pressure decrease from the inlet to the outlet of the vessel.  It is given by:

${\Delta}P\ =\frac{128{\mu}{L}}{\pi{d^4}}{Q}$

where ${\Delta}P$ is the pressure decrease, ${\mu}$ is the viscosity of blood, L is the length of the vessel, and d is the diameter of the vessel.

You wish to explore how the pressure drop changes across the arteries when arteriosclerosis begins to occur, meaning that the radius of the artery becomes smaller.

Use viscosity = ${4 x 10^{-3}}$ Pa-s for whole blood at 37 degC. Use ${Q_{tot}=80 \frac{cm^3}{sec}}$ and $Q = \frac{Q_{tot}}{N}$, where N is the number of vessels in a section, given in the table below. Note: 1 Pascal (Pa) = 0.00750062 mmHg.  The table below shows pertinent values for all sections of the vasculature.

# Action Potentials
We can use computer modeling to describe the behavior of neuronal action potentials in various conditions. A very simple neuron model represents the neuron cell membrane as a resistor-capacitor (RC) circuit, similar how we modeled neurons in class.

As we studied in class and observed in our cricket and locust labs, neurons tend to remain at a resting voltage potential until they are externally stimulated above some threshold, which causes their voltage potential to rapidly spike (the action potential) before returning to their resting potential.

In our model here, the stimulus to our model neuron will come in the form of injected current.  If sufficient current is injected to bring the membrane potential above threshold, the neuron should produce a spike.

Here we will build a model that establishes a base membrane potential, allows for injection of a stimulus current, fires a spike if the membrane potential rises above a threshold, then resets the membrane potential afterward.

(As you know from our lab work and other course work, this wouldn't be a terribly good model, since we have to code in the reset part instead of the model compensating naturally like the voltage-gated ion channels, but it's not that bad.)




We showed in module 2 that the membrane can be modeled by the equivalent resistance of all of its ion channels and the membrance capacitance, which represents the stored charge across the membrane at equilibrium. For example, from Lecture 6:



By Kirkhoff's Law, we know that the sum of currents entering and leaving a node is zero.  Therefore our current through the membrane can be described as an external current ${I_e}$ as follows:

${I_e} ={I_R}+ {I_C}$

We know that ${I_R}$ is the current through the resistive elements, meaning the ion channels.  By Ohm's Law (${V = IR}$), we know that ${I_R}$ will be equal to ${g_m(V_m-V_{rest})}$, where ${g_m}$ is the conductance (1/R) through the ion channels, and $(V_m-V_{rest})$ is the difference between the membrane potential at a given moment and the equilibrium or resting membrane potential.

From our studies of capacitance, we know that the current through the capacitor is given by:
${I_C} = {C_m}\frac{d{V_m}}{dt}$.
If we put this all together, we get

${I_e} ={I_R}+ {I_C}$

${I_e} ={g_m(V_m-V_{rest})}+ {C_m}\frac{d{V_m}}{dt}$

or

${C_m}\frac{d{V_m}}{dt} = {I_e} - {g_m(V_m-V_{rest})}$

It's the membrane voltage {V_m} that we're interested in, so let's solve for it.

Multiplying both sides by R gives:

$R{C_m}\frac{d{V_m}}{dt} = R{I_e} - {V_m+V_{rest}}$

using ${\tau} = R{C_m}$,

$\frac{d{V_m}}{dt} = {\frac{1}{\tau}}R{I_e} - {V_m+V_{rest}}$

At any particular time point, you know from calculus that this first order differential equation has the general solution

${V_m(t)} = {V_{m,\infty}}+(V(0)-{V_{m,\infty}}){e^{\frac{-dt}{\tau}}}$          <---- this is the main equation

where V(0) refers to the value of ${V_m}$ at the previous time point.

What is ${V_{m,\infty}}$? (I'm just going to call it ${V_\infty}$ from here.) From our previous work in the class, we know from the ion Nernst potentials that the resting membrane potential is about -70mV. So when there is no injected current, ${V_m(t)}$ = ${V_\infty}$ = -70mV for all time t.

Let's practice with this model.

## Set up the model

We will solve this model in a step-wise format, recalculating the value of $V_m$ at every time point, based on the previous value of $V_m$.

You can assume the following:

$V_{rest} = -70 mV$

$R_m = 10M\Omega$, the membrane resistance

$\tau = 10ms$, the membrane time constant

$V_{threshold} = -55 mV$, the voltage level that triggers an action potential

$V_{reset} = -75 mV$, the voltage level that the membrane returns to after an action potential

$V_{spike} = 20 mV$, the voltage level that the membrane potential reaches at the peak of an action potential

dt = 0.05 ms, the size of our time step

Add code to the box below to initialize these six variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# add your code here to initialize the variables
# you should keep the units shown above. Do not convert any units.
# For example, set Rm = 10, not Rm = 10000000, and tau = 10, not tau = 0.01

#constants
Vrest= -70
Rm = 10
Tau= 10
Vthreshold= -55
Vreset= -75
Vspike= 20
dt= 0.05
```

Next, we'll set up the vectors to hold our data.  Run the following code block:

```python
# Creates a vector of time points from 0 to 499 ms in steps of dt=0.05
timeVector = np.arange(0, 500, dt)

# Creates a placeholder vector for our membrane voltages that is the same size as timeVector
VmVector = np.zeros(len(timeVector))

# Creates a placeholder for the external stimulation vector (the injected current).
# It is also the same size as the time vector.
stimVector = np.zeros(len(timeVector))

# Set the initial voltage to be equal to the resting potential
VmVector[0] = Vrest # use your own variable name here if it's different than Vrest
```

In the next cell, we'll set the injected current. Let's start with an injected current of 0.5 nA for the first 100 ms. Run the following code block:

```python
stimVector[0:2000] = 0.5 # this sets the injected current to 0.5 nA for the first 100 ms of the simulation
```

Here's your job now: to write the main part of the model:

Write a *for loop* that steps a variable S through every time step.

*Hint: since we've already calculated the value of Vm at the first time step (VmVector[0]), set your for loop to run for one less iteration than the length of the time vector.)*

Inside the for loop, do the following:
* calculate a fresh value for Vinf, where Vinf = Vrest + Rm*(stimVector[S])
* calculate a new value for Vm, VmVector[S+1] = Vinf + (VmVector[S] - Vinf) * np.exp(-dt/tau)
* write an *if* statement that checks to see if the new value of Vm is greater than or equal to the threshold voltage
* if it is, set that new value of Vm to be equal to Vspike
* inside that same *if* statement, check to see if the previous value of Vm (VmVector[S]) was already at Vspike
* if it was, set the new value of Vm (VmVector[S+1]) to be equal to Vreset

Each of the above actions is one line of code.  You should therefore have six lines of code inside your for loop (not counting any comments you may have).

```python
# I'll start you off
for S in range(len(timeVector)-1):
  Vinf = Vrest + Rm*(stimVector[S])
  VmVector[S+1] = Vinf + (VmVector[S] - Vinf) * np.exp(-dt/Tau)

if VmVector[S+1] >= Vthreshold:  
        VmVector[S+1] = Vspike  
        if VmVector[S] == Vspike:
            VmVector[S+1] = Vreset

```

Finally, plot the Vm voltage vector over time.  As an example, your first plot should look like this:


```python
import numpy as np
import matplotlib.pyplot as plt

Vrest = -70
Rm = 10
Tau = 10
Vthreshold = -55
Vreset = -75
Vspike = 20
dt = 0.05

timeVector = np.arange(0, 500, dt)
VmVector = np.zeros(len(timeVector))
stimVector = np.zeros(len(timeVector))

VmVector[0] = Vrest
stimVector[0:2000] = 0.5  

for S in range(len(timeVector)-1):
    Vinf = Vrest + Rm * stimVector[S]
    VmVector[S+1] = Vinf + (VmVector[S] - Vinf) * np.exp(-dt / Tau)

    if VmVector[S+1] >= Vthreshold:
        VmVector[S+1] = Vspike
        if VmVector[S] == Vspike:
            VmVector[S+1] = Vreset

plt.plot(timeVector, VmVector)
plt.title('Voltage versus time')
plt.xlabel('Time in ms')
plt.ylabel('Voltage in mV')
plt.grid(True)
plt.legend()
plt.show()
```

As the levels of stimulus input current increase, the maximum Voltage in mV also increases. At 0.5, the maximum is at -65. At 1.3, the maximum is -57. At 2.0, the maximum is 20.

```python
import numpy as np
import matplotlib.pyplot as plt

Vrest = -70
Rm = 10
Tau = 10
Vthreshold = -55
Vreset = -75
Vspike = 20
dt = 0.05

timeVector = np.arange(0, 500, dt)
VmVector = np.zeros(len(timeVector))
stimVector = np.zeros(len(timeVector))

VmVector[0] = Vrest
stimVector[0:2000] = 1.6


for S in range(len(timeVector)-1):
    Vinf = Vrest + Rm * stimVector[S]
    VmVector[S+1] = Vinf + (VmVector[S] - Vinf) * np.exp(-dt / Tau)

    if VmVector[S+1] >= Vthreshold:
        VmVector[S+1] = Vspike
        if VmVector[S] == Vspike:
            VmVector[S+1] = Vreset

plt.plot(timeVector, VmVector)
plt.title('Voltage versus time')
plt.xlabel('Time in ms')
plt.ylabel('Voltage in mV')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

Vrest = -70
Rm = 10
Tau = 10
Vthreshold = -55
Vreset = -75
Vspike = 20
dt = 0.05

timeVector = np.arange(0, 500, dt)
VmVector = np.zeros(len(timeVector))
stimVector = np.zeros(len(timeVector))

VmVector[0] = Vrest
stimVector[0:2000] = 1.0
stimVector[2500:4000] = 2.6
stimVector[5000:7000] = 4.0

for S in range(len(timeVector)-1):
    Vinf = Vrest + Rm * stimVector[S]
    VmVector[S+1] = Vinf + (VmVector[S] - Vinf) * np.exp(-dt / Tau)

    if VmVector[S+1] >= Vthreshold:
        VmVector[S+1] = Vspike
        if VmVector[S] == Vspike:
            VmVector[S+1] = Vreset

plt.plot(timeVector, VmVector)
plt.title('Voltage versus time')
plt.xlabel('Time in ms')
plt.ylabel('Voltage in mV')
plt.grid(True)
plt.legend()
plt.show()
```
